```python
import pandas as pd
```

```python
news=pd.read_csv('inte.csv')
```

```python
news
```

```python
news['DATE_TIME']=news['DATE_TIME'].str.split('_').str[0]
```

```python
news['NEWS_NUM']=range(1,332257)
```

```python
split=news['ITEM_NAME'].str.split('|').apply(lambda x: pd.Series(x)).stack().reset_index(level=1, drop=True).to_frame('ITEM_NAME_ONLY')
```

```python
news=news.merge(split, left_index=True, right_index=True, how='left')
```

```python
news
```

```python
news=news.reset_index(drop=True)
```

```python
news=news.drop('ITEM_NAME', axis=1)
```

```python
news
```
